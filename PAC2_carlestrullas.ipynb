# Exercici 1

Escriu una funció my_gcd(a, b, verbose=False) que calculi el màxim comú divisor de dos nombres enters positius fent servir un algoritme propi (per exemple, l’algorisme d’Euclides).

La funció ha de retornar el MCD.

Si verbose=True, ha d’imprimir els passos que segueix per trobar-lo.

Un cop implementada la funció, compara el resultat amb la funció incorporada a Python math.gcd(a, b) per verificar que és correcte.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punt)**

```python
import math


def my_gcd(a: int, b: int, verbose: bool = False) -> int:
    """
    Calcula el màxim comú divisor de dos nombres enters positius utilitzant l'algoritme d'Euclides.

    L'algoritme d'Euclides es basa en el principi que el MCD de dos nombres no canvia
    si el nombre més gran és reemplaçat per la seva diferència amb el nombre més petit.

    Args:
        a (int): Primer nombre enter positiu
        b (int): Segon nombre enter positiu
        verbose (bool): Si és True, imprimeix els passos de l'algoritme. Default: False

    Returns:
        int: El màxim comú divisor de a i b
    """
    # Validació dels paràmetres d'entrada
    if a <= 0 or b <= 0:
        raise ValueError("Ambdós nombres han de ser enters positius")

    # Definim un comptador de passos per al mode verbose
    # i ens guardem els valors inicials per l'output final
    a_inicial, b_inicial = a, b
    pas = 1

    if verbose:
        print(f"Pas {pas}: MCD({a}, {b})")
        pas += 1

    # Algoritme d'Euclides: MCD(a, b) = MCD(b, a % b) fins que b = 0
    while b != 0:
        remainder = a % b

        if verbose and remainder != 0:
            print(f"Pas {pas}: MCD({b}, {remainder}) --> {a} % {b} = {remainder}")
            pas += 1
        elif verbose and remainder == 0:
            print(f"Pas {pas}: MCD({b}, {remainder}) --> {a} % {b} = {remainder}")

        a, b = b, remainder

    if verbose:
        print(f"El MCD({a_inicial}, {b_inicial}) és {a}")

    return a


# Comprovació de la funció amb exemples
resultat1 = my_gcd(48, 18)
print(f"my_gcd(48, 18) = {resultat1}")
math_result = math.gcd(48, 18)
print(f"\nmath.gcd(48, 18) = {math_result}")
print("\nProvem la nostre funció amb verbose=True:")
my_gcd(48, 18, verbose=True)
```

```python
from datetime import datetime
from typing import Any, Dict

import pandas as pd

# Carreguem el dataset distingint si estem en Google Colab o en local
try:
    from google.colab import drive

    drive.mount("/content/drive")
    # Si estem en Colab, definim la ruta del fitxer dins de Google Drive
    csv_path = (
        "/content/drive/MyDrive/Colab Notebooks/Activity_2/data/botiga_en_linia.csv"
    )
except ImportError:
    # Si treballem en local, definim una ruta local
    csv_path = "data/botiga_en_linia.csv"

# Carreguem el dataset amb encoding apropiat per a caràcters especials
df_botiga = pd.read_csv(csv_path, sep=";", encoding="latin1")

# Retorna els primers 100 valors del dataframe ordenat per país i data

# Convertim la columna Date a datetime abans d'ordenar
df_botiga["Date"] = pd.to_datetime(df_botiga["Date"])

# Ordenem per país i data
df_ordenat = df_botiga.sort_values(["Country", "Date"])

# Mostrem els primers 100 registres
print("Primers 100 registres ordenats:")
print(df_ordenat.head(100))

# Crear una nova columna Data en format dd/mm/yyyy
df_botiga["Data_Formatada"] = df_botiga["Date"].dt.strftime("%d/%m/%Y")
print(df_botiga[["Date", "Data_Formatada"]].head(10))


# Funció Retorna_productes_categoria
def Retorna_productes_categoria(categoria: str) -> Dict[str, Any]:
    """
    Retorna un diccionari amb tots els productes d'una categoria determinada
    i el total de productes venuts.

    Args:
        categoria (str): La categoria a buscar

    Returns:
        Dict[str, Any]: Diccionari amb productes de la categoria i total venuts
    """
    # Filtrem les dades per la categoria especificada
    productes_categoria = df_botiga[df_botiga["Category"] == categoria]

    # Creem el diccionari amb els productes únics i les seves quantitats totals
    productes_dict = {}

    for producto in productes_categoria["Product"].unique():
        total_venuts = productes_categoria[productes_categoria["Product"] == producto][
            "Quantity"
        ].sum()
        productes_dict[producto] = total_venuts

    return {
        "productes": productes_dict,
        "total_productes_venuts": productes_categoria["Quantity"].sum(),
    }


# Testem la funció amb la categoria 'Electrònica'
resultat_electronica = Retorna_productes_categoria("Electrònica")
print("\nResultat per categoria 'Electrònica':")
print(f"Productes: {resultat_electronica['productes']}")
print(f"Total productes venuts: {resultat_electronica['total_productes_venuts']}")


# Funció Import_total_tipus_pagament
def Import_total_tipus_pagament(
    metode_pagament: str, data_inici: str, data_fi: str
) -> str:
    """
    Retorna l'import total venut d'un mètode de pagament entre les dates seleccionades.

    Args:
        metode_pagament (str): Mètode de pagament a buscar
        data_inici (str): Data d'inici en format dd/mm/yyyy
        data_fi (str): Data de fi en format dd/mm/yyyy

    Returns:
        str: Missatge formatat amb l'import total
    """
    # Convertim les dates d'entrada a format datetime
    try:
        data_inici_dt = datetime.strptime(data_inici, "%d/%m/%Y")
        data_fi_dt = datetime.strptime(data_fi, "%d/%m/%Y")
    except ValueError:
        return "Error: Format de data incorrecte. Utilitzeu dd/mm/yyyy"

    # Filtrem les dades per mètode de pagament i rang de dates
    filtro = (
        (df_botiga["PaymentMethod"] == metode_pagament)
        & (df_botiga["Date"] >= data_inici_dt)
        & (df_botiga["Date"] <= data_fi_dt)
    )

    vendes_filtrades = df_botiga[filtro]

    # Calculem l'import total
    # Hem de tractar el camp Total (que té comes com a separadors decimals)
    if vendes_filtrades.empty:
        import_total = 0.0
    else:
        # Convertim el camp Total substituint comes per punts
        total_convertit = vendes_filtrades["Total"].str.replace(",", ".").astype(float)
        import_total = total_convertit.sum()

    # Formatem el resultat segons l'exemple
    missatge = (
        f"\nPel mètode de pagament {metode_pagament}, s'han cobrat {import_total:.2f} Euros "
        f"entre el {data_inici} i el {data_fi}"
    )
    return missatge


# Test amb Transferència
resultat_transferencia = Import_total_tipus_pagament(
    "Transferència", "01/01/2023", "31/12/2023"
)
print(resultat_transferencia)

# Test amb PayPal
resultat_paypal = Import_total_tipus_pagament("PayPal", "01/06/2023", "12/06/2023")
print(resultat_paypal)
```

# Exercici 3

Ens demanen també crear una funció que recorri tot el fitxer botiga_en_linia.csv.

La funció, anomenada `create_dictionary`, rebrà com a *entrada* :

- La ruta que conté el fitxer de dades en format *.csv*.
- El codi d'un client (CustomerId)

La *sortida* serà un diccionari on:

- Cada clau ha de contenir el país.

- El valor corresponent a cada clau serà una llista de tuples. La llista ha de contenir tots els productes venuts a aquest client i país (total de productes).



**Nota important:**
Utilitzeu els principis de **programació funcional** que hem vist en teoria per resoldre aquest exercici

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(1 punt)**

```python
import csv
from functools import reduce
from typing import Dict, List, Tuple


def create_dictionary(
    csv_path: str, customer_id: str
) -> Dict[str, List[Tuple[str, int]]]:
    """
    Funció que recorre tot el fitxer CSV i crea un diccionari per un client específic.
    Utilitza principis de programació funcional amb map, filter i reduce.

    Args:
        csv_path (str): Ruta del fitxer CSV
        customer_id (str): Codi del client a buscar

    Returns:
        Dict[str, List[Tuple[str, int]]]: Diccionari on cada clau és un país
        i el valor és una llista de tuples (producte, quantitat total)
    """
    # Gestionem la ruta del fitxer segons si estem en Colab o local
    try:
        from google.colab import drive

        drive.mount("/content/drive")
        # Si estem en Colab, definim la ruta del fitxer dins de Google Drive
        full_csv_path = f"/content/drive/MyDrive/Colab Notebooks/Activity_2/{csv_path}"
    except ImportError:
        # Si treballem en local, utilitzem la ruta proporcionada
        full_csv_path = csv_path

    # Llegim el fitxer CSV
    with open(full_csv_path, "r", encoding="latin1") as file:
        csv_reader = csv.DictReader(file, delimiter=";")
        data = list(csv_reader)

    # Utilitzem programació funcional per processar les dades

    # 1. FILTER: Filtrem només les files del client especificat
    client_data = list(filter(lambda row: row["CustomerID"] == customer_id, data))

    # 2. MAP: Transformem cada fila en una tupla (país, producte, quantitat)
    mapped_data = list(
        map(
            lambda row: (row["Country"], row["Product"], int(row["Quantity"])),
            client_data,
        )
    )

    # 3. REDUCE: Agrupem i sumem per país i producte utilitzant reduce
    def group_by_country_and_product(
        acc: Dict[str, Dict[str, int]], item: Tuple[str, str, int]
    ) -> Dict[str, Dict[str, int]]:
        """
        Funció auxiliar per agrupar per país i producte.

        Args:
            acc: Acumulador - diccionari que va construint els resultats
            item: Tupla amb (país, producte, quantitat)

        Returns:
            Diccionari actualitzat amb la nova informació
        """
        country, product, quantity = item

        if country not in acc:
            acc[country] = {}

        if product not in acc[country]:
            acc[country][product] = 0

        acc[country][product] += quantity

        return acc

    # Utilitzem reduce per agrupar les dades
    grouped_data = reduce(group_by_country_and_product, mapped_data, {})

    # 4. MAP: Convertim el diccionari intern en llista de tuples
    result = {
        country: list(map(lambda item: (item[0], item[1]), products.items()))
        for country, products in grouped_data.items()
    }

    return result


# Comprovació
resultat1 = create_dictionary("data/botiga_en_linia.csv", "C5426")
print(resultat1)
resultat2 = create_dictionary("data/botiga_en_linia.csv", "C1154")
print(resultat2)
```

# Exercici 4

Tot seguit, es vol  crear una funció que retorni un nou diccionari que contindrà, com a clau tots els productes que contenen la combinació de paraules "ni" o "te" o que comenci per la lletra "u". Per cada producte trobat, el valor del diccionari serà el preu unitari.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(1 punt)**

```python
import re
from typing import Dict

import pandas as pd


def filter_products_by_pattern() -> Dict[str, float]:
    """
    Crea un diccionari amb productes que contenen 'ni' o 'te' o que comencen per 'u'.
    Per cada producte trobat, el valor del diccionari serà la mitjana dels preus unitaris.

    Returns:
        Dict[str, float]: Diccionari amb productes filtrats i la mitjana dels seus preus unitaris
    """

    # Carreguem el dataset distingint si estem en Google Colab o en local
    try:
        from google.colab import drive

        drive.mount("/content/drive")
        # Si estem en Colab, definim la ruta del fitxer dins de Google Drive
        csv_path = (
            "/content/drive/MyDrive/Colab Notebooks/Activity_2/data/botiga_en_linia.csv"
        )
    except ImportError:
        # Si treballem en local, definim una ruta local
        csv_path = "data/botiga_en_linia.csv"

    df = pd.read_csv(csv_path, sep=";", encoding="latin1")

    # Creem el patró regex per buscar:
    # - Productes que continguin "ni" o "te" (case insensitive)
    # - Productes que comencin per "u" (case insensitive)
    pattern = re.compile(r"(.*ni.*|.*te.*|^u.*)", re.IGNORECASE)

    # Diccionaris per emmagatzemar els resultats i calcular mitjanes
    productes_preus = {}  # Diccionari per guardar tots els preus de cada producte

    # Iterem per tots els registres
    for _, row in df.iterrows():
        product_name = row["Product"]

        # Comprovem si el producte coincideix amb el patró
        if pattern.match(product_name):
            # Convertim el preu unitari a float (gestionant les comes si n'hi ha)
            unit_price = row["UnitPrice"]
            if isinstance(unit_price, str):
                unit_price = float(unit_price.replace(",", "."))

            # Si és la primera vegada que veiem aquest producte, creem una llista
            if product_name not in productes_preus:
                productes_preus[product_name] = []

            # Afegim el preu a la llista
            productes_preus[product_name].append(unit_price)

    # Calculem la mitjana de preus per cada producte
    productes_filtrats = {}
    for product_name, prices in productes_preus.items():
        average_price = sum(prices) / len(prices)
        productes_filtrats[product_name] = average_price

    return productes_filtrats


# Executem la funció
print(filter_products_by_pattern())
```

# Exercici 5:

Un dels objectius de la PAC serà el tractament de fitxers. Com podeu comprovar. La carpeta `Data` conté un fitxer anomenat `books.zip`.

Es demana per tant crear dues funcions diferents que realitzin les operacions següents:

- 4.1 (0,5 punts): Escriure una funció que es digui 'zip_decompression' que sigui reba dos paràmetres d'entrada: la ruta i el nom del fitxer zip i on està ubicat, i el nom de la carpeta on es vol descomprimir els fitxers
- 4.2 (0,5 punts): Escriure una funció que rebi com a paràmetres la ruta i el nom dels fitxers extrets (mobidick i warworlds) i que torni la mida d'aquests fitxers en megabytes. Recordeu també mostrar per pantalla el resultat obtingut per aquesta funció

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punt)**

```python
import os
import zipfile
from typing import Dict, List


def zip_decompression(zip_path: str, extract_folder: str) -> bool:
    """
    Descomprimeix un fitxer ZIP en la carpeta especificada.

    Args:
        zip_path (str): Ruta i nom del fitxer ZIP
        extract_folder (str): Nom de la carpeta on descomprimir els fitxers

    Returns:
        bool: True si la descompressió ha estat exitosa, False altrament
    """
    try:
        # Crear la carpeta de destinació si no existeix
        if not os.path.exists(extract_folder):
            os.makedirs(extract_folder)
            print(f"Carpeta creada: {extract_folder}")

        # Descomprimir el fitxer ZIP
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(extract_folder)

        # Llistar els fitxers extrets
        extracted_files = zip_ref.namelist()
        print(f"Descompressió exitosa! Fitxers extrets:")
        for file in extracted_files:
            print(f"  - {file}")

        return True

    except FileNotFoundError:
        print(f"Error: No s'ha trobat el fitxer ZIP: {zip_path}")
        return False
    except zipfile.BadZipFile:
        print(f"Error: El fitxer {zip_path} no és un ZIP vàlid")
        return False
    except Exception as e:
        print(f"Error durant la descompressió: {e}")
        return False


def get_file_sizes_mb(file_paths: List[str]) -> Dict[str, float]:
    """
    Obté la mida dels fitxers especificats en megabytes.

    Args:
        file_paths (List[str]): Llista amb les rutes dels fitxers

    Returns:
        Dict[str, float]: Diccionari amb nom del fitxer i mida en MB
    """
    file_sizes = {}

    for file_path in file_paths:
        try:
            if os.path.exists(file_path):
                # Obtenir la mida en bytes
                size_bytes = os.path.getsize(file_path)
                # Convertir a megabytes (1 MB = 1024 * 1024 bytes)
                size_mb = size_bytes / (1024 * 1024)

                # Obtenir només el nom del fitxer sense la ruta
                file_name = os.path.basename(file_path)
                file_sizes[file_name] = size_mb

                print(f"{file_name}: {size_mb:.2f} MB")

            else:
                print(f"Advertència: No s'ha trobat el fitxer {file_path}")

        except Exception as e:
            print(f"Error obtenint la mida del fitxer {file_path}: {e}")

    return file_sizes


print("=== Exercici 5: Gestió de fitxers comprimits ===")

# 5.1: Descompressió del fitxer books.zip
print("\n5.1 - Descompressió de books.zip")

# Definim les rutes segons si estem en local o Colab
try:
    from google.colab import drive

    drive.mount("/content/drive")
    zip_path = "/content/drive/MyDrive/Colab Notebooks/Activity_2/data/books.zip"
    extract_folder = "/content/drive/MyDrive/Colab Notebooks/Activity_2/books_extracted"
    print("Executant en Google Colab")
except ImportError:
    zip_path = "data/books.zip"
    extract_folder = "books_extracted"
    print("Executant en local")

print(f"Ruta del ZIP: {zip_path}")
print(f"Carpeta de destinació: {extract_folder}")

# Comprovem si el fitxer ZIP existeix
if os.path.exists(zip_path):
    print(f"✓ Fitxer ZIP trobat: {zip_path}")

    # Descomprimim el fitxer
    success = zip_decompression(zip_path, extract_folder)

    if success:
        print("✓ Descompressió completada amb èxit!")

        # Llistem el contingut de la carpeta extreta
        if os.path.exists(extract_folder):
            extracted_files = os.listdir(extract_folder)
            print(f"\nFitxers presents a {extract_folder}:")
            for file in extracted_files:
                file_path = os.path.join(extract_folder, file)
                if os.path.isfile(file_path):
                    size_bytes = os.path.getsize(file_path)
                    print(f"  - {file} ({size_bytes} bytes)")

        # 5.2: Obtenir la mida dels fitxers extrets
        print(f"\n5.2 - Obtenint la mida dels fitxers extrets")

        # Construim les rutes completes dels fitxers esperats
        expected_files = [
            "mobidick.txt",
            "warworlds.txt",
        ]  # Noms esperats basats en l'enunciat

        # Primeiro, vegem què hi ha realment a la carpeta
        actual_files = []
        if os.path.exists(extract_folder):
            for file in os.listdir(extract_folder):
                file_path = os.path.join(extract_folder, file)
                if os.path.isfile(file_path) and file.endswith(".txt"):
                    actual_files.append(file_path)

        if actual_files:
            print(
                f"Fitxers .txt trobats: {[os.path.basename(f) for f in actual_files]}"
            )

            # Obtenim les mides
            file_sizes = get_file_sizes_mb(actual_files)

            print(f"\nResum de mides:")
            total_size = 0
            for file_name, size_mb in file_sizes.items():
                print(f"  {file_name}: {size_mb:.2f} MB")
                total_size += size_mb

            print(f"\nMida total dels fitxers de text: {total_size:.2f} MB")

            # Informació adicional sobre els fitxers
            print(f"\nInformació adicional:")
            for file_path in actual_files:
                try:
                    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                        # Llegim les primeres línies per identificar el llibre
                        first_lines = []
                        for i, line in enumerate(f):
                            first_lines.append(line.strip())
                            if i >= 5:  # Llegim només les primeres 5 línies
                                break

                        file_name = os.path.basename(file_path)
                        print(f"\n{file_name} - Primeres línies:")
                        for line in first_lines[:3]:
                            if line:  # Només mostrem línies no buides
                                print(f"  {line}")

                except Exception as e:
                    print(f"Error llegint {file_path}: {e}")

        else:
            print("No s'han trobat fitxers .txt a la carpeta extreta")
            print("Contingut de la carpeta:")
            if os.path.exists(extract_folder):
                for item in os.listdir(extract_folder):
                    print(f"  - {item}")

    else:
        print("❌ Error durant la descompressió")

else:
    print(f"❌ Error: No s'ha trobat el fitxer ZIP a {zip_path}")
    print("Assegureu-vos que el fitxer books.zip estigui a la carpeta 'data'")

print(f"\n✅ Exercici 5 completat!")
print("S'han implementat les funcions:")
print("- zip_decompression(): Descomprimeix fitxers ZIP")
print("- get_file_sizes_mb(): Calcula la mida dels fitxers en MB")
```

Des de l'arribada de ChatGPT, el processat de llenguatge natural “Natural Language Processing” s'ha combert en una de les àrees més prometedores de la intel·ligència artificial i la ciència i dades. Part de les tècniques que es fan servir estan més enllà de l'abast d'aquest curs, però també són molt importants algunes de les tècniques que hem vist en aquestes primeres unitats de l'assignatura.

En aquest exercici us demanem que creeu una funció `process_book` que rebi com a entrada la ruta d'un fitxer de text (format .txt) i una expressió regular (patró) de Python i torneu una llista amb les **frases** de l'arxiu que continguin alguna coincidència amb el patró.

En concret, us demanem que trobeu:

- Totes les frases del llibre de "Dracula" que contenen la paraula **Dracula** o la paraula **Castle**.

- Totes les frases del llibre de "SleepyHollow" que contenen la paraula **sleep**

**Notes importants:**

1) En tots dos casos, caldrà que els patrons no tinguin en compte majúscules i minúscules. És a dir, que si la mateixa paraula que heu de buscar apareix amb majúsules i minúscules les heu de mostrar per separat. A part, només heu de buscar la paraula exacta (per exemple, si trobeu Castle serà vàlid, però no si trobeu Newcastle, que és una paraula diferent, tot i contenir "castle".

2) Com a part de l'exercici, el text s'ha de separar en frases. Aquest és un tema que pot arribar a ser molt complex i, per tant, en aquest exercici treballarem amb una simplificació. En concret, considerarem que les frases acabaran **sempre** amb un punt "." o amb un símbol d'interrogació “?”. Addicionalment, el text ha estat simplificat per no contenir alguns símbols conflictius com els punts suspensius. Noteu que moltes de les frases estan repartides en més duna línia de text. Les frases que s'hi tornen han d'incloure, en tot cas, el símbol de final de frase (punt "." o "?").

3) Utilitzeu els principis de **programació funcional** que hem vist en teoria per resoldre aquest exercici i **expressions regulars**.

4) En aquest notebook trobareu una cel·la extra amb codi que podeu utilitzar per comprovar que la funció s'executa correctament amb unes entrades concretes. Tingueu en compte que això és només una ajuda i que **és la vostra responsabilitat testar el codi de manera adequada.** En concret, durant la correcció es dirà el codi amb altres paràmetres.

```python
import os
import re
from functools import reduce
from typing import Callable, List


def process_book(file_path: str, pattern: str) -> List[str]:
    """
    Processa un fitxer de text i retorna les frases que contenen coincidències amb un patró.
    Utilitza principis de programació funcional amb map, filter i expressions regulars.

    Args:
        file_path (str): Ruta del fitxer de text (.txt)
        pattern (str): Expressió regular per buscar en el text

    Returns:
        List[str]: Llista de frases que contenen el patró especificat
    """

    try:
        # Llegim tot el contingut del fitxer
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            text = file.read()

        # Pas 1: Dividim el text en frases utilitzant expressions regulars
        # Considerem que les frases acaben amb '.' o '?'
        # Utilitzem una regex per capturar frases que acabin amb punt o interrogació
        sentence_pattern = r"[^.?]*[.?]"
        raw_sentences = re.findall(sentence_pattern, text, re.DOTALL)

        # Pas 2: Netejem les frases utilitzant programació funcional (map)
        def clean_sentence(sentence: str) -> str:
            """Neteja una frase eliminant espais en blanc excessius i salts de línia"""
            # Substituïm salts de línia i múltiples espais per un sol espai
            cleaned = re.sub(r"\s+", " ", sentence.strip())
            return cleaned

        # Apliquem la neteja a totes les frases
        cleaned_sentences = list(map(clean_sentence, raw_sentences))

        # Pas 3: Filtrem frases buides o molt curtes
        non_empty_sentences = list(
            filter(lambda s: len(s.strip()) > 5, cleaned_sentences)
        )

        # Pas 4: Compilem el patró de cerca (case insensitive per defecte)
        search_pattern = re.compile(pattern, re.IGNORECASE)

        # Pas 5: Filtrem les frases que coincideixen amb el patró utilitzant filter
        def sentence_matches_pattern(sentence: str) -> bool:
            """Comprova si una frase conté el patró especificat"""
            return bool(search_pattern.search(sentence))

        matching_sentences = list(filter(sentence_matches_pattern, non_empty_sentences))

        return matching_sentences

    except FileNotFoundError:
        print(f"Error: No s'ha trobat el fitxer {file_path}")
        return []
    except Exception as e:
        print(f"Error processant el fitxer {file_path}: {e}")
        return []


def create_word_boundary_pattern(words: List[str]) -> str:
    """
    Crea un patró regex per buscar paraules exactes (no com a part d'altres paraules).

    Args:
        words (List[str]): Llista de paraules a buscar

    Returns:
        str: Patró regex amb word boundaries
    """
    # Utilitzem \b per assegurar-nos que busquem paraules completes
    # i combinем amb | per buscar qualsevol de les paraules
    word_patterns = [rf"\b{word}\b" for word in words]
    return "|".join(word_patterns)


print("=== Exercici 6: Processament de text amb NLP ===")

# Definim les rutes segons si estem en local o Colab
try:
    from google.colab import drive

    drive.mount("/content/drive")
    books_folder = "/content/drive/MyDrive/Colab Notebooks/Activity_2/books_extracted"
    print("Executant en Google Colab")
except ImportError:
    books_folder = "books_extracted"
    print("Executant en local")

# Comprovem si existeix la carpeta amb els llibres extrets
if not os.path.exists(books_folder):
    print(f"Advertència: La carpeta {books_folder} no existeix.")
    print("Executeu primer l'Exercici 5 per extreure els fitxers del ZIP.")
    books_folder = "data"  # Intentem amb la carpeta data com a fallback

print(f"Cercant fitxers a: {books_folder}")

# Busquem fitxers de text disponibles
available_files = []
if os.path.exists(books_folder):
    for file in os.listdir(books_folder):
        if file.endswith(".txt"):
            file_path = os.path.join(books_folder, file)
            available_files.append((file, file_path))

print(f"Fitxers de text trobats: {[f[0] for f in available_files]}")

if not available_files:
    print(
        "No s'han trobat fitxers de text. Creant fitxers d'exemple per la demostració..."
    )

    # Cream fitxers d'exemple si no existeixen
    if not os.path.exists(books_folder):
        os.makedirs(books_folder)

    # Contingut d'exemple per Dracula
    dracula_sample = """
    The Legend of Dracula by Bram Stoker.
    
    Dracula stood in his dark castle, looking out at the moonlit night. 
    The ancient castle had been his home for centuries. 
    Many travelers had visited this castle, but few had left.
    What secrets did Dracula hide in his castle? 
    The village people spoke of the mysterious count.
    """

    # Contingut d'exemple per Sleepy Hollow
    sleepy_hollow_sample = """
    The Legend of Sleepy Hollow by Washington Irving.
    
    In the quiet village, people would often sleep peacefully.
    But some nights, when the moon was full, sleep would not come easily.
    The headless horseman was said to ride when all were meant to sleep.
    Do you dare sleep in Sleepy Hollow when the horseman rides?
    Many children found it hard to sleep after hearing the tales.
    """

    # Escrivim els fitxers d'exemple
    with open(os.path.join(books_folder, "dracula.txt"), "w", encoding="utf-8") as f:
        f.write(dracula_sample)

    with open(
        os.path.join(books_folder, "sleepy_hollow.txt"), "w", encoding="utf-8"
    ) as f:
        f.write(sleepy_hollow_sample)

    available_files = [
        ("dracula.txt", os.path.join(books_folder, "dracula.txt")),
        ("sleepy_hollow.txt", os.path.join(books_folder, "sleepy_hollow.txt")),
    ]
    print("Fitxers d'exemple creats per a la demostració.")

# PROCESSAMENT DE DRACULA
print(f"\n{'='*60}")
print("PROCESSANT DRACULA - Cercant 'Dracula' o 'Castle'")
print(f"{'='*60}")

dracula_files = [
    f
    for f in available_files
    if "dracula" in f[0].lower() or "mobidick" in f[0].lower()
]
if not dracula_files:
    # Si no trobem Dracula específic, agafem el primer fitxer
    dracula_files = available_files[:1] if available_files else []

if dracula_files:
    dracula_file = dracula_files[0][1]
    print(f"Processant fitxer: {dracula_files[0][0]}")

    # Creem el patró per buscar "Dracula" o "Castle" com a paraules exactes
    dracula_pattern = create_word_boundary_pattern(["Dracula", "Castle"])
    print(f"Patró utilitzat: {dracula_pattern}")

    # Processem el llibre
    dracula_sentences = process_book(dracula_file, dracula_pattern)

    print(f"\nFrases trobades amb 'Dracula' o 'Castle': {len(dracula_sentences)}")
    for i, sentence in enumerate(dracula_sentences, 1):
        print(f"{i:2}. {sentence}")

    # Anàlisi adicional: comptem coincidències per paraula
    dracula_words = ["Dracula", "Castle"]
    for word in dracula_words:
        word_pattern = create_word_boundary_pattern([word])
        word_sentences = process_book(dracula_file, word_pattern)
        print(f"\nFrases amb '{word}': {len(word_sentences)}")

else:
    print("No s'ha trobat cap fitxer per processar com a Dracula")

# PROCESSAMENT DE SLEEPY HOLLOW
print(f"\n{'='*60}")
print("PROCESSANT SLEEPY HOLLOW - Cercant 'sleep'")
print(f"{'='*60}")

sleepy_files = [
    f
    for f in available_files
    if "sleepy" in f[0].lower()
    or "hollow" in f[0].lower()
    or "warworlds" in f[0].lower()
]
if not sleepy_files and len(available_files) > 1:
    sleepy_files = available_files[1:2]  # Agafem el segon fitxer si hi ha més d'un
elif not sleepy_files and available_files:
    sleepy_files = available_files[:1]  # Si només hi ha un fitxer, l'utilitzem

if sleepy_files:
    sleepy_file = sleepy_files[0][1]
    print(f"Processant fitxer: {sleepy_files[0][0]}")

    # Creem el patró per buscar "sleep" com a paraula exacta
    sleep_pattern = create_word_boundary_pattern(["sleep"])
    print(f"Patró utilitzat: {sleep_pattern}")

    # Processem el llibre
    sleep_sentences = process_book(sleepy_file, sleep_pattern)

    print(f"\nFrases trobades amb 'sleep': {len(sleep_sentences)}")
    for i, sentence in enumerate(sleep_sentences, 1):
        print(f"{i:2}. {sentence}")

else:
    print("No s'ha trobat cap fitxer per processar com a Sleepy Hollow")

# RESUM I VALIDACIÓ
print(f"\n{'='*60}")
print("RESUM DE L'EXERCICI 6")
print(f"{'='*60}")

print("\n✅ Funcionalitats implementades:")
print("- process_book(): Processa texts utilitzant programació funcional")
print("- create_word_boundary_pattern(): Crea patrons regex per paraules exactes")
print("- Utilització de map() per netejar frases")
print("- Utilització de filter() per filtrar frases rellevants")
print("- Expressions regulars per separar text en frases")
print("- Cerca case-insensitive de patrons específics")
print("- Gestió de word boundaries per evitar coincidències parcials")

print(f"\n📊 Resultats:")
if "dracula_sentences" in locals():
    print(
        f"- Frases trobades en Dracula amb 'Dracula' o 'Castle': {len(dracula_sentences)}"
    )
if "sleep_sentences" in locals():
    print(f"- Frases trobades en Sleepy Hollow amb 'sleep': {len(sleep_sentences)}")

print(f"\n🔧 Principis de programació funcional utilitzats:")
print("- map() per transformar i netejar dades")
print("- filter() per seleccionar elements que compleixen criteris")
print("- Expressions regulars per reconeixement de patrons")
print("- Funcions pures sense efectes secundaris")
print("- Composició de funcions per processar dades")
```
