En aquesta sèrie d’exercicis treballarem l’eficiència temporal dels algorismes.
Analitzarem, optimitzarem i compararem diferents implementacions d’un problema senzill: detectar si una llista conté elements duplicats.

L’objectiu és entendre com l’elecció d’un enfocament o un altre pot modificar de manera notable el rendiment d’un algorisme.
Recordeu que el rendiment d’un algorisme no depèn només de la mida de l’entrada, sinó també dels seus valors específics. Per aquest motiu, quan s’analitza l’eficiència, habitualment es considera el cas pitjor, és a dir, aquell en què l’algorisme necessita el nombre més elevat d’operacions per completar-se.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **Exercici 1.1 (0.5 punts)**

En aquest exercici heu d’analitzar la complexitat temporal del codi següent i expressar-ne l’ordre de complexitat.

A partir de l’algorisme següent, determineu-ne la complexitat en notació O, justificant breument la vostra resposta.

````python
def has_duplicates(num_list: list[int]) -> bool:
    """
    Checks whether a list of ints contains duplicate values.

    Args:
        num_list (list[int]): List of integers to analyze.

    Returns:
        bool: True if duplicates are found, False otherwise.
    """
    for k in range(len(num_list)):
        for v in range(k + 1, len(num_list)):
            if num_list[k] == num_list[v]:
                return True
    return False
````

**Resposta:**

L’algorisme té una complexitat $\Theta(n^2)$, ja que executa un bucle dins d’un altre per comparar tots els parells possibles d’elements dins de la llista.
Concretament, el primer bucle (k) recorre tots els elements de la llista, i per a cadascun d’ells, el segon bucle (v) compara aquell element amb tots els que el segueixen.
Com que la llista té una mida n, el nombre total de comparacions realitzades és aproximadament:

$$
(n - 1) + (n - 2) + (n - 3) + \dots + 1 = \frac{n(n - 1)}{2}
$$

Aquest valor creix de manera proporcional a $n^2$, i per tant la complexitat temporal és quadràtica, $\Theta(n^2)$.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **Exercici 1.2 (0.75 punts)**

En aquest exercici heu de millorar la funció `has_duplicates` de manera que en redueixi la complexitat temporal. Per fer-ho, creeu una nova funció anomenada `has_duplicates_improved` i implementeu-hi el codi corresponent.

Comproveu que totes dues versions retornen el mateix resultat quan s’executen amb diferents valors d’entrada. Per fer-ho, afegiu una verificació addicional utilitzant `assert`. Un cop millorada, indiqueu la complexitat temporal del nou algorisme i expliqueu breument en què consisteix la millora respecte de la versió original.

````python
def has_duplicates_improved(lst: list[int]) -> bool:
    """
    Checks whether a list of ints contains duplicate values using a set.

    Args:
        lst (list[int]): List of integers to analyze.

    Returns:
        bool: True if duplicates are found, False otherwise.
    """
    duplicated = set()
    for k in lst:
        if k in duplicated:
            return True
        duplicated.add(k)
    return False
````

**Resposta:**

Aquesta versió recorre la llista una sola vegada, de manera que la seva complexitat temporal és $\Theta(n)$.
Per aconseguir aquesta millora, ha sigut necessari utilitzar una estructura de dades addicional, en aquest cas un conjunt (set), que permet comprovar si un element ja ha aparegut en temps constant mitjà, O(1), gràcies al mecanisme de hashing intern que utilitzen els conjunts.

Tot i que una complexitat $\Theta(n)$ ja es pot considerar relativament lenta quan les dades són molt grans, continua sent molt més eficient que una complexitat $\Theta(n^2)$.
En el cas d’un algorisme quadràtic, el temps d’execució creixeria molt més ràpidament, fent-lo inviable per a llistes de gran mida.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **Exercici 1.3 (0.75 punts)**

En aquest exercici heu de comparar el temps d’execució dels dos algorismes que verifiquen si una llista conté elements duplicats: l’algorisme donat i la vostra versió millorada.

- Genereu llistes d’aquests tamanys amb elements sense repetir i mesureu el temps d’execució de tots dos algorismes per a cada tamany.

- Representeu els resultats en una gràfica que mostri com creix el temps d’execució en funció del tamany de la llista (n).

- Comenteu les diferències observades entre els dos algorismes, relacionant-les amb les seves complexitats teòriques.


Utilitzeu els següents tamanys de llista:

```python
sizes = [200, 600, 1000, 1400, 1800, 5000]
```


````python
import random
import time

import matplotlib.pyplot as plt

sizes = [200, 600, 1000, 1400, 1800, 5000]
times_slow, times_fast = [], []

for n in sizes:
    data = list(range(n))
    random.shuffle(data)

    start = time.time()
    has_duplicates(data)
    times_slow.append(time.time() - start)

    start = time.time()
    has_duplicates_improved(data)
    times_fast.append(time.time() - start)

plt.plot(sizes, times_slow, 'o-', label='has_duplicates (O(n²))')
plt.plot(sizes, times_fast, 'o-', label='has_duplicates_improved (O(n))')
plt.xlabel('n (List size)')
plt.ylabel('Time (s)')
plt.title('Execution time growth comparison')
plt.legend()
plt.grid(True)
plt.show()
````

**Resposta:**

La corba de l’algorisme donat mostra un creixement exponencial del temps d’execució a mesura que augmenta la mida de la llista. És a dir, quan n creix, el temps es dispara ràpidament perquè l’algorisme realitza comparacions entre molts parells d’elements.

En canvi, la línia taronja (has_duplicates_improved, amb complexitat $\Theta(n)$) es manté pràcticament constant fins i tot amb llistes grans. Això indica que l’ús d’un conjunt (set) permet fer les comprovacions de duplicats de manera molt més eficient, ja que cada cerca és gairebé instantània.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 2.1 (0.75 punts)**

Heu d’implementar dues funcions que calculin el temps mitjà de reproducció segons el tipus de subscripció (`subscription_type`). Disposeu de la columna `listening_time`.

- Implementació A: realitzeu la lectura del fitxer de manera manual utilitzant les funcions integrades de Python (open() i read()), processant línia per línia les dades per calcular el promig.

- Implementació B: implementeu la mateixa funcionalitat utilitzant la llibreria pandas, aprofitant els seus mètodes (read_csv, groupby i mean).

Les dues implementacions han de de rebre únicament el paràmetre `path: str`, corresponent a la ruta del fitxer de dades. En ambdós casos, no s’ha de fer cap manipulació addicional ni mostrar informació per pantalla. Les funcions han de retornar únicament un diccionari amb el resultat calculat.

````python
def avg_listening_time_manual(path: str) -> dict:
    """
    Calculates the average 'listening_time' per 'subscription_type'
    using manual file reading (open + read).
    
    Args:
    path (str): Path to the CSV file.

    Returns:
        dict: keys are subscription types.
    """
    with open(path, "r") as f:
        # Find column idx
        header = [h.strip().lower() for h in f.readline().strip().split(",")]
        sub_idx = header.index("subscription_type")
        play_idx = header.index("listening_time")

        sums, counts = {}, {}
        for line in f:
            parts = line.strip().split(",")
            sub = parts[sub_idx]
            play = float(parts[play_idx])

            # Accumulate total listening time and count
            sums[sub] = sums.get(sub, 0.0) + play
            counts[sub] = counts.get(sub, 0) + 1

    return {k: sums[k] / counts[k] for k in sums}
````

````python
import pandas as pd

def avg_listening_time_pandas(path: str) -> dict:
    """
    Calculates the average 'listening_time' per 'subscription_type'
    using pandas (optimized reading and group by operations).

    Args:
        path (str): Path to the CSV file.

    Returns:
        dict: Keys are subscription types and are avg listening time.
    """
    df = pd.read_csv(path)
    result = df.groupby("subscription_type")["listening_time"].mean()
    return result.to_dict()
````

## Exercici 3

En aquest exercici detectareu les ineficiències i millorareu el codi. En aquest cas, analitzarem el càlcul de la suma dels factorials dels primers **n** nombres naturals.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **Exercici 3.1 (1 punt)**

Utilitzant les eines de *profiling* que hem vist al notebook de teoria, analitzeu el fragment de codi següent i determineu on es troba el coll d’ampolla. Expliqueu les ineficiències que heu detectat, relacionant-les amb els resultats obtinguts en el *profiling*.

````python
def factorial_sum(n: int):
    """Returns the sum of the factorials of the first n natural numbers.

    Args:
        n (int): Positive integer indicating the upper limit of the factorial sum.

    Returns:
        (int): Sum of factorials from 1! to n!.
    """
    total = 0
    for i in range(1, n + 1):
        fact = 1
        for j in range(1, i + 1):
            fact *= j
        total += fact
    return total

# factorial_sum(3500)
````

````python
%lprun -f factorial_sum factorial_sum(3500)
````

**Resposta:**

Si utilitzem el profiler, el primer que observem és que la funció triga aproximadament 7,8 segons (segons l’equip), i que el percentatge més gran del temps d’execució es concentra en el bucle intern, on es recalcula el factorial complet per a cada nombre. Això indica que la principal ineficiència prové de repetir operacions que ja s’havien fet en iteracions anteriors.

Per a cada valor de i, es tornava a calcular el factorial des de 1, cosa que generava un cost quadràtic (O(n²)) a causa del bucle imbricat. Per tant, l’optimització d’aquesta funció consisteix a evitar el càlcul repetit dels factorials dins del bucle.
  

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 3.2 (0.75 punts)**

Milloreu el codi amb l’objectiu de reduir el temps d’execució. A més, comenteu la millora realitzada, indicant-ne la complexitat temporal i si aquesta ha millorat respecte de la versió anterior.

````python
def factorial_sum_optimized(n: int) -> int:
    """Optimized, computes the sum of factorials from 1! to n!.
    
    Args:
        n (int): Positive integer indicating the upper limit of the factorial sum.

    Returns:
        int: Sum of factorials from 1! to n!.
    """
    total: int = 0
    fact: int = 1
    for i in range(1, n + 1):
        fact *= i
        total += fact
    return total

# factorial_sum_optimized(3500)
````

````python
%lprun -f factorial_sum_optimized factorial_sum_optimized(3500)
````

**Resposta:**

El factorial es reutilitza d’una iteració a l’altra, multiplicant simplement el resultat anterior pel nou valor de i. D’aquesta manera, cada factorial s’obté amb una única operació addicional, reduint la complexitat temporal a O(n). Aquesta millora fa que l’algorisme sigui molt més eficient, especialment per a valors grans de n, mantenint el mateix resultat final.

````python
def run_sequential(nums: list[int]):
    """
    Computes the factorial of each number sequentially 
    and prints the total execution time.

    Args:
        nums (list[int]): List of ints to process.
    """
    print("Sequential mode:")
    start = time.time()
    results = [calcular_factorial(n) for n in nums]
    end = time.time()
    print(f"Total sequential time: {end - start:.4f}s")
````

````python
import multiprocessing

def run_multiprocessing(nums: list[int]):
    """
    Computes the factorial of each number using multiprocessing 
    and prints the total execution time.

    Args:
        nums (list[int]): List of ints to process.
    """
    print("Multiprocessing mode:")
    start = time.time()
    with multiprocessing.Pool() as pool:
        results = pool.map(calcular_factorial, nums)
    end = time.time()
    print(f"Total multiprocessing time: {end - start:.4f}s")
````

Una empresa dedicada a la creació de continguts digitals us ha sol·licitat desenvolupar una galeria d’imatges obtingudes des d’una API pública. L’objectiu del projecte és descarregar les imatges i emmagatzemar-les localment per al seu posterior ús dins d’una aplicació interna de gestió i visualització.

En una primera fase del desenvolupament, s’han recopilat diverses URLs d’imatges i es vol avaluar l’eficiència del procés de descàrrega sota diferents enfocaments d’execució.

Per a això, podeu accedir al llistat d’imatges a descarregar cridant la funció `get_image_urls()`, que retorna una llista d’URLs d’una API pública. A més, es proporciona un exemple de funció per descarregar una imatge individual, suposant que ja existeix la carpeta `imgs/` on s’emmagatzemaran els fitxers descarregats.

````python
import requests

def get_image_urls(num: int = 50) -> list[str]:
    """
    Fetches a list of random image URLs from the API.

    Args:
        num (int): Number of image URLs to retrieve.

    Returns:
        list[str]: A list containing the image URLs.
    """
    api_url = f"https://picsum.photos/v2/list?page=1&limit={num}"
    response = requests.get(api_url)
    return [item["download_url"] for item in response.json()]
````

````python
import os

def download_image(url: str, idx: int) -> None:
    """
    Downloads a single image from a given URL and 
    saves it with a unique index.

    Args:
        url (str): The URL of the image to download.
        idx (int): The index used to create a unique filename.
    """
    filename = os.path.join("imgs", f"image_{idx}.jpg")
    try:
        r = requests.get(url)
        with open(filename, "wb") as f:
            f.write(r.content)
        print(f"Saved: {filename}")
    except requests.RequestException as e:
        print(f"Error downloading {url}: {e}")
````

Haureu d’implementar una funció que recorri la llista d’URLs i descarregui totes les imatges una a una, cridant la funció download_image(url, idx) per a cada element.
Durant el procés, heu de mesurar i mostrar el temps total d’execució, de manera que es pugui conèixer quant tarda el procés seqüencial complet.

**Nota:**

Abans de començar la descàrrega, assegureu-vos que existeix la carpeta on s’emmagatzemaran les imatges (per exemple, imgs/).
Per a això, es recomana utilitzar la funció `os.makedirs()` amb el paràmetre `exist_ok=True`, que permet crear la carpeta automàticament en cas que no existeixi.

````python
import time

def download_images_sequential(urls: list[str]) -> None:
    """
    Downloads all images sequentially and measures execution time.

    Args:
        urls (list[str]): List of image URLs to download.
    """
    os.makedirs("imgs", exist_ok=True)
    print("\n Starting sequential download...")

    start_time = time.time()

    for idx, url in enumerate(urls, start=1):
        download_image(url, idx)

    total_time = time.time() - start_time
    print(f"\n Sequential download completed in {total_time:.2f} seconds.")
````

**Nota:**

Utilitzeu les classes `ThreadPoolExecutor` i `ProcessPoolExecutor` per facilitar l’execució de codi en múltiples threads i processos.

````python
import os
import shutil
import matplotlib.pyplot as plt


def prepare_images_folder(folder_name: str = "imgs") -> None:
    """
    Deletes the existing folder (if any) and recreates it empty.
    """
    if os.path.exists(folder_name):
        shutil.rmtree(folder_name)
    os.makedirs(folder_name, exist_ok=True)


def measure_time(func, *args, **kwargs) -> float:
    """Utility function to measure execution time of a given function."""
    start = time.time()
    func(*args, **kwargs)
    return time.time() - start


def plot_results(results: dict, thread_n_process_counts: list[int]) -> None:
    """
    Plots a comparison graph showing execution times for sequential,
    multithreading, and multiprocessing download methods.

    Args:
        results (dict): Dictionary containing execution times for each method.
        thread_counts (list[int]): List of thread counts tested.
        process_counts (list[int]): List of process counts tested.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(thread_n_process_counts, results["multithreading"], marker="o", label="Multithreading")
    plt.plot(thread_n_process_counts, results["multiprocessing"], marker="o", label="Multiprocessing")
    plt.axhline(y=results["sequential"][0], color="gray", linestyle="--", label="Sequential (base)")
    plt.title("Comparison of download times (50 images)")
    plt.xlabel("Number of threads / processes")
    plt.ylabel("Total time (seconds)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()
````

````python
import concurrent.futures
import requests


def download_images_multithreaded(urls: list[str], num_of_threads: int = 5) -> None:
    """
    Downloads multiple images using a ThreadPoolExecutor.
    Prints thread names and measures total execution time.
    """
    print(f"\n Starting multithreaded download with {num_of_threads} threads...\n")

    with concurrent.futures.ThreadPoolExecutor(max_workers=num_of_threads) as executor:
        futures = [executor.submit(download_image, url, idx) for idx, url in enumerate(urls, start=1)]
        for future in concurrent.futures.as_completed(futures):
            future.result()
````

````python
urls = get_image_urls(num=50)

# Store times
results = {
    "sequential": [],
    "multithreading": [],
    "multiprocessing": [],
}
thread_n_process_counts = [1, 2, 4, 6, 8, 10]
````

````python
# Sequential
print("Sequential mode")
prepare_images_folder()
t_seq = measure_time(download_images_sequential, urls)
results["sequential"].append(t_seq)
print(f"Total sequential time: {t_seq:.2f} s\n")
````

````python
# Multithreading
for threads in thread_n_process_counts:
    print(f"Multithreading mode ({threads} threads)")
    prepare_images_folder()
    t = measure_time(download_images_multithreaded, urls, threads)
    results["multithreading"].append(t)
    print(f"Total time with {threads} threads: {t:.2f} s\n")
````
