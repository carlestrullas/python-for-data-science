# Rendiment dels estudiants universitaris a Catalunya

## Descripció del Dataset: Rendiment dels estudiants universitaris a Catalunya

El conjunt de dades recull els principals indicadors vinculats a la mesura del rendiment acadèmic dels estudiants, com són les taxes de rendiment, de graduació, d'eficiència i d'abandonament. Les dades han estat descarregades des del següent enllaç: 
* https://recercaiuniversitats.gencat.cat/ca/03_ambits_dactuacio/xifres/docencia/Rendiment-dels-estudiants/


Es treballarà amb dos datasets principals:

### 1. Taxa de rendiment (rendiment_estudiants.xlsx)
Per a un determinat curs, és el quocient entre el nombre de crèdits superats respecte al nombre de crèdits matriculats en els estudis de graus i màsters.

* En aquest càlcul només es tenen en compte els **crèdits ordinaris**, que són els propis d'un estudi.
* S'exclouen els crèdits reconeguts, els adaptats i els transferits d'altres plans d'estudis.

### 2. Taxa d'abandonament (taxa_abandonament.xlsx)
Per a un determinat curs, és la proporció d'estudiants de grau o màster que abandonen el Sistema Universitari Català (SUC) en el primer curs sobre el total de nou accés d'aquell curs.

* S'estableix que l'estudiant ha abandonat el SUC si no es matricula en el proper curs en cap de les universitats catalanes.

# Exercici 4: Anàlisi estadística automatitzada


En aquest exercici, crearàs una funció anomenada `analyze_dataset()` que s'encarregarà de realitzar una anàlisi estadística completa del dataset fusionat que vas obtenir a l'exercici 2. L'objectiu és calcular diferents mètriques i estadístiques que ens permetin entendre millor el comportament del rendiment i l'abandonament universitari, i desar tots aquests resultats en un arxiu JSON ben estructurat.

**Estructura de l'anàlisi JSON:**

L'arxiu JSON que has de generar (`src/report/analisi_estadistic.json`) ha de contenir els següents apartats. Observa com està organitzat en seccions lògiques que van del general (metadata i estadístiques globals) a l'específic (anàlisi per branca). Podràs veure un exemple de l'estructura a (`examples/analisi_estadistic_example.json`) (*Nota: Els valors de l'exemple no són els correctes*).

### 4.1. Secció Metadata:
   
   Aquesta secció ha de contenir informació bàsica sobre l'anàlisi que estàs realitzant. Pensa en ella com la "fitxa tècnica" de la teva anàlisi:
   
   - **fecha_analisis**: La data actual en format ISO (YYYY-MM-DD). Utilitza `datetime.now().strftime("%Y-%m-%d")` per obtenir-la automàticament.
   - **num_registros**: El nombre total de registres al teu dataset fusionat. Simplement utilitza `len(merged_df)`.
   - **periodo_temporal**: Una llista ordenada amb tots els cursos acadèmics únics que apareixen a les teves dades. Pots obtenir-la amb `sorted(merged_df['Curs Acadèmic'].unique())`.

### 4.2. Estadístiques Globals:
   
   Aquí calcularàs les mètriques que resumeixen el comportament general de tot el sistema universitari català, sense diferenciar per branques:
   
   - **abandono_medio**: La taxa mitjana d'abandonament en primer curs, calculada com la mitjana de la columna `'% Abandonament a primer curs'`.
   - **rendimiento_medio**: La taxa mitjana de rendiment, calculada com la mitjana de la columna `'Taxa rendiment'`.
   - **correlacion_abandono_rendimiento**: Aquest és un valor molt interessant que ens indica si existeix relació entre l'abandonament i el rendiment. Per calcular-lo, has d'utilitzar la correlació de Pearson de scipy:
   
   ```python
   from scipy.stats import pearsonr
   
   corr, p_value = pearsonr(
       merged_df['% Abandonament a primer curs'].dropna(),
       merged_df['Taxa rendiment'].dropna()
   )
   ```
   
   Una correlació negativa (com -0.68) ens indica que a major abandonament, menor rendiment, la qual cosa té sentit intuitiu.

### 4.3. Anàlisi per Branca:
   
   Aquesta és la secció més completa de l'anàlisi. Per a cada branca d'estudis (Arts i humanitats, Ciències, Ciències de la salut, etc.), has de calcular:
   
   - **Estadístiques descriptives bàsiques:**
     - Mitjana i desviació estàndard del percentatge d'abandonament
     - Mitjana i desviació estàndard de la taxa de rendiment
   
   - **Detecció de tendències temporals:**
     
     Aquí és on aplicaràs la regressió lineal per detectar si l'abandonament en cada branca està millorant (tendència decreixent), empitjorant (tendència creixent), o es manté estable en el temps.
     
     El procés és el següent:
     
     a) Primer, agrupa les dades per any acadèmic per a cada branca:
     ```python
     branch_data = merged_df[merged_df['Branca'] == branch]
     branch_by_year = branch_data.groupby('Curs Acadèmic').agg({
         '% Abandonament a primer curs': 'mean'
     }).reset_index()
     ```
     
     b) Extreu les llistes d'anys i valors:
     ```python
     years = branch_by_year['Curs Acadèmic'].tolist()
     valores_abandono = branch_by_year['% Abandonament a primer curs'].tolist()
     ```
     
     c) Calcula la regressió lineal utilitzant scipy:
     ```python
     from scipy.stats import linregress
     
     slope, intercept, r_value, p_value, std_err = linregress(
         range(len(years)),  # Posicions: 0, 1, 2, 3...
         valores_abandono
     )
     ```
     
     d) Interpreta el pendent (slope):
     - Si `slope > 0.01`: la tendència és "creixent" (l'abandonament augmenta amb el temps)
     - Si `slope < -0.01`: la tendència és "decreixent" (l'abandonament disminueix)
     - Si està entre -0.01 i 0.01: la tendència és "estable" (sense canvis significatius)

### 4.4. Rankings:
