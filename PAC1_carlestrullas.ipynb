## Exercici 1

Carregueu el *dataset* `employees.xlsx` de manera que obtingueu un objecte de tipus `DataFrame`.

Una vegada carregat el *dataset*, resol les següents qüestions:

1. Mostrar el nombre total d'empleats en el DataFrame.
2. Calcular el salari màxim, mínim i mitjà de tots els empleats.
3. Mostrar quants empleats pertanyen a cada departament (pista: value_counts()).
4. Crear una nova columna Age que contingui l'edat de cada empleat (a partir de *DateOfBirth*).
5. Mostrar els 5 empleats amb major salari.
6. Mostrar quants empleats tenen un cognom que comença per la lletra "S".
7. Calcular la mitjana de salaris dels empleats del departament "IT".
8. Crear una nova columna Initials que contingui les inicials de cada empleat (exemple: "John Smith" → "J.S.").

**Pista 1:** Recordeu que podeu utilitzar llibreries com a `pandas` per a carregar el conjunt de dades de manera senzilla i obtenir un `DataFrame`.

**Pista 2:** El *type hint* d'un `DataFrame` és `pd.DataFrame`.

**Tots els exercicis s'han de resoldre a partir del `DataFrame` obtingut en finalitzar aquest exercici i no carregar-lo cada vegada que s'hagi d'utilitzar.**

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 punts)**

```python
# Carreguem imports necessaris, ordenats segons PEP8
from collections import OrderedDict
from datetime import datetime

import pandas as pd

# Carreguem el dataset distingint si estem en Google Colab o en local
try:
    from google.colab import drive

    drive.mount("/content/drive")
    # Si estem en Colab, definim la ruta del fitxer dins de Google Drive
    file_path = "/MyDrive/Colab Notebooks/Activity 1/data/employees.xlsx"
except ImportError:
    # Si treballem en local, definim una ruta local. Assumim que el fitxer
    # .xlsx està a la carpeta 'data' dins del directori on es troba el notebook
    file_path = "data/employees.xlsx"

# Carreguem el dataset creant un DataFrame de pandas
df_employees: pd.DataFrame = pd.read_excel(file_path)

# 1. Calculem el nombre total d'empleats amb la funció len()
print(f"El nombre total d'empleats és: {len(df_employees)}")

# 2. Calculem el salari màxim, mínim i mitjà amb max(), min() i mean()
print("\nSalari màxim:", df_employees["Salary"].max())
print("Salari mínim:", df_employees["Salary"].min())
print("Salari mitjà:", df_employees["Salary"].mean())

# 3. Calculem el nombre d'empleats per departament amb value_counts()
print("\nNombre d'empleats a cada departament:")
print(df_employees["Department"].value_counts())

# 4. Definim una nova columna "Age" calculant l'edat a partir de "DateOfBirth"
# amb una funció lambda. Mostrem els 5 primers registres.
df_employees["Age"] = df_employees["DateOfBirth"].apply(
    lambda date_of_birth: datetime.now().year - pd.to_datetime(date_of_birth).year
)
print("\nMostrem els 5 primers registres:")
print(df_employees[["FirstName", "LastName", "Age", "DateOfBirth"]].head(5))

# 5. Mostrem els 5 empleats amb major salari ordenant i aplicant head(5)
print("\nTop 5 empleats amb major salari:")
print(
    df_employees[["FirstName", "LastName", "Salary"]]
    .sort_values(by="Salary", ascending=False)
    .head(5)
)

# 6. Comptem els cognoms que comencen amb 'S' amb la funció str.startswith()
print(
    "\nNombre d'empleats amb cognom que comença per 'S':",
    df_employees["LastName"].str.startswith("S").sum(),
)

# 7. Calculem la mitjana de salaris dels empleats del departament "IT"
mitjana_it = df_employees[df_employees["Department"] == "IT"]["Salary"].mean()
print(f"\nLa mitjana de salaris dels empleats del departament 'IT' és: {mitjana_it}")

# 8. Creem una nova columna Initials que contingui les inicials de cada empleat
df_employees["Initials"] = (
    df_employees["FirstName"].str[0] + "." + df_employees["LastName"].str[0] + "."
)
print("\nMostrem els 5 primers registres amb les inicials:")
print(df_employees[["FirstName", "LastName", "Initials"]].head(5))
```

## Exercici 2

Utilitzant el dataframe obtingut en l'Exercici 1:

1. Calcular el salari mitjà per departament i mostrar quin departament té la mitjana més alta.

2. Comptar quants empleats hi ha a cada departament i mostrar el resultat en ordre descendent.

3. Llistar els noms i cognoms de les persones el salari de les quals estigui per sobre del salari mig general.


<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 punts)**

```python
# 1. Salari mitjà per departament i departament amb mitjana més alta. Bàsicament
# utilitzem groupby() i mean() per agrupar per departament i calcular la mitjana
# de salaris. Després, amb idxmax() obtenim el departament amb la mitjana més alta.
salary_by_department = df_employees.groupby("Department")["Salary"].mean()
print("Salari mitjà per departament:")
print(salary_by_department)
print("\nDepartament amb la mitjana més alta:", salary_by_department.idxmax())

# 2. Comptem el nombre d'empleats per departament amb value_counts(), que ja
# ho torna de forma descendent.
employees_by_department = df_employees["Department"].value_counts()
print("\n Nombre d'empleats per departament en ordre descendent:")
print(employees_by_department)

# 3. Noms i cognoms amb salari per sobre de la mitjana. Per fer-ho, calculem la
# mitjana de salaris i filtrem el DataFrame per obtenir només els empleats amb
# salari superior a aquesta mitjana.
average_salary = df_employees["Salary"].mean()
above_average = df_employees[df_employees["Salary"] > average_salary]
print("\nEmpleats amb salari per sobre de la mitjana:")
print(above_average[["FirstName", "LastName", "Salary"]])
```

## Exercici 3

Utilitzant el dataframe obtingut en l'Exercici 1:

1. Crear la funció **rang_empleats** que rebi com a paràmetres el DataFrame i un rang d'edats (min_edat i max_edat).

2. La funció ha de calcular l'edat de cada empleat a partir del seu *DateOfBirth*.

3. La funció ha de retornar un diccionari amb:

"comptatge" → nombre total d'empleats dins del rang d'edat.

"salari_mig" → salari mitjà d'aquests empleats.

Exemple:

<pre>
resultat = rang_empleats(df, min_edat=30, max_edat=50)
print(resultat)

{'comptatge': 44, 'salari_mig': 53174.53}
</pre>

Nota:

Es pot calcular l'edat realitzant la diferència entre l'any actual i l'any de *DateOfBirth*

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 punts)**

```python
# Definim la funció rang_empleats
def rang_empleats(df: pd.DataFrame, min_edat: int, max_edat: int) -> dict:
    if max_edat <= min_edat:
        raise ValueError("max_edat ha de ser estrictament més gran que min_edat")

    current_year = datetime.now().year
    df["Age"] = df["DateOfBirth"].apply(
        lambda date_of_birth: current_year - pd.to_datetime(date_of_birth).year
    )
    filtered = df[(df["Age"] >= min_edat) & (df["Age"] <= max_edat)]
    return {
        "comptatge": len(filtered),
        "salari_mig": round(filtered["Salary"].mean(), 2),
    }


# Comprovem la funció rang_empleats
try:
    resultat = rang_empleats(df_employees, min_edat=30, max_edat=50)
    print(resultat)
except ValueError as e:
    print(e)
```

## Exercici 4

1. Donada una llista de números de telèfon, heu de recórrer-la i, aplicant expressions regulars, determinar si el número és vàlid o no vàlid.

Un número vàlid ha de tenir el format:

- Opcional +34 o 0034 a l'inici (codi de país).
- Seguit de 9 dígits.
- Pot tenir espais, guions o punts com a separadors entre els dígits.
- Exemples vàlids: +34 612-345-678, 0034.612.345.678, 612345678.

2. Donat un text que conté hores, heu d'extreure, a través d'una única expressió regular, una llista amb totes les hores que tinguin el format correcte:

- Format correcte: HH:MM o HH:MM:SS
- HH: 0–23
- MM i SS: 0–59
- Exemple vàlid: 08.30, 23.59:59, 0:5:7 (permetent 1 o 2 dígits).

Podeu utilitzar per a validar l'exercici els següents exemples d'ús:

<pre>
# Llista de números de telèfon
numero_movil = [
"+34 612-345-678",
"0034.612.345.678",
"612345678",
"+34 612 345 67", # invàlid
"12345678",       # invàlid
"+341234567890"   # invàlid
]

# Extracció d'hores

text_amb_temps = """
La reunió comença a les 08:30 i acaba a les 17:45:30.
Una altra hora errònia: 25:61, i una altra correcta: 0:5:7.
"""
</pre>
<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 punts)**

```python
import re

numero_movil = [
    "+34 612-345-678",
    "0034.612.345.678",
    "612345678",
    "+34 612 345 67",  # invàlid
    "12345678",  # invàlid
    "+341234567890",  # invàlid
]

valid_phone_pattern = re.compile(r"^(\+34|0034)?[ .-]?\d{3}[ .-]?\d{3}[ .-]?\d{3}$")
# Explicació de la regex que he construit:
# ^: Marca l'inici de la cadena.
# (\+34|0034)?: Grup opcional que permet el codi de país espanyol:
#     - \+34: Accepta el codi de país amb un signe '+' seguit de '34'.
#     - 0034: Accepta el codi de país amb '0034' com a alternativa.
#     - ?: Fa que aquest grup sigui opcional (pot aparèixer o no).
# [ .-]?: Permet un separador opcional (espai, punt o guió) després del codi de país.
# \d{3}: Accepta exactament 3 dígits.
# [ .-]?: Permet un altre separador opcional (espai, punt o guió).
# \d{3}: Accepta exactament 3 dígits més.
# [ .-]?: Permet un altre separador opcional (espai, punt o guió).
# \d{3}: Accepta exactament 3 dígits més.
# $: Marca el final de la cadena.

for numero in numero_movil:
    if valid_phone_pattern.match(numero):
        print(f"{numero} és vàlid")
    else:
        print(f"{numero} és invàlid")

text_amb_temps = """
La reunió comença a les 08:30 i acaba a les 17:45:30.
Una altra hora errònia: 25:61, i una altra correcta: 0:5:7.
"""
valid_time_pattern = re.compile(
    r"\b(?:[01]?\d|2[0-3]):(?:[0-5]?\d)(?::(?:[0-5]?\d))?\b"
)
# Explicació de la regex que he construit:
# \b: Marca un límit de paraula. Això assegura que la coincidència comenci i
# acabi en els límits d'una paraula, evitant que es trobin coincidències
# parcials dins d'altres cadenes.

# (?:[01]?\d|2[0-3]):
# - Aquest és un grup no capturat (indicat per (?: ... )) que especifica les hores (HH):
#   - [01]?\d: Permet hores de 0 a 19.
#       - [01]?: Permet opcionalment un "0" o un "1" al principi.
#       - \d: Accepta qualsevol dígit (0-9).
#   - |: Alternativa lògica (o).
#   - 2[0-3]: Permet hores de 20 a 23.
# - El ":" literal separa les hores dels minuts.

# (?:[0-5]?\d)
# - Aquest és un altre grup no capturat que especifica els minuts (MM):
#   - [0-5]?: Permet opcionalment un dígit entre 0 i 5.
#   - \d: Accepta qualsevol dígit (0-9).

# (?::(?:[0-5]?\d))?
# - Aquest grup és opcional (indicat pel "?" final) i especifica els segons (SS):
#   - ":" Literal que separa els minuts dels segons.
#   - (?:[0-5]?\d): Grup no capturat que especifica els segons:
#       - [0-5]?: Permet opcionalment un dígit entre 0 i 5.
#       - \d: Accepta qualsevol dígit (0-9).

# \b: Marca el final del límit de paraula.

# En resum, aquesta regex captura hores en format HH:MM o HH:MM:SS, on:
# - HH (hores) pot ser un valor entre 0 i 23.
# - MM (minuts) pot ser un valor entre 0 i 59.
# - SS (segons) és opcional i pot ser un valor entre 0 i 59.
valid_times = valid_time_pattern.findall(text_amb_temps)
print("\nHores vàlides trobades:", valid_times)
```

## Exercici 5

Utilitzant el dataframe obtingut en l'Exercici 1:

1. Crea una nova columna anomenada **Categoria_Treball** que contingui la categoria del treball segons la columna *JobTitle*.

Per exemple:
- Si *JobTitle* conté "Engineer" → "Engineering"
- Si *JobTitle* conté "Manager" → "Management"
- Si *JobTitle* conté "Teacher" → "Education"
- Si no compleix cap de les anteriors → "Other"

2. Crea un diccionari **category_counts** que contingui el nombre d'empleats per cada *JobCategory* usant dict comprehension.

3. Ordena el diccionari per nombre d'empleats descendent i guarda'l com *OrderedDict*.

4. Imprimeix el diccionari recorrent-lo amb un bucle for.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 punts)**

```python
# 1. Afegim la columna Categoria_Treball
def categorize_job(title: str) -> str:
    if "Engineer" in title:
        return "Engineering"
    elif "Manager" in title:
        return "Management"
    elif "Teacher" in title:
        return "Education"
    else:
        return "Other"


# Apliquem la funció per crear la nova columna
df_employees["Categoria_Treball"] = df_employees["JobTitle"].apply(categorize_job)

# 2. Crear diccionari category_counts usant dict comprehension
category_counts = {
    category: df_employees[df_employees["Categoria_Treball"] == category].shape[0]
    for category in df_employees["Categoria_Treball"].unique()
}
print(f"Dict sense ordenar: {category_counts}")

# 3. Ordenem el diccionari
ordered_category_counts = OrderedDict(
    sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
)
print(f"\nDict ordenat (ordered dict): {ordered_category_counts}")

# 4. Imprimim el diccionari
print("\nNombre d'empleats per categoria de treball:")
for category, count in ordered_category_counts.items():
    print(f"{category}: {count}")
```
